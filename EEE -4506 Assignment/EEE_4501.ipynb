```python
import numpy as np
import matplotlib.pyplot as plt

#  
x = np.arange(0, 10.1, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(-x)

#  
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

#  subplot (Sin(x))
axs[0, 0].plot(x, y1)
axs[0, 0].set_title('Sin(x)')

#  subplot (Cos(x))
axs[0, 1].plot(x, y2)
axs[0, 1].set_title('Cos(x)')

#  subplot (Tan(x))
axs[1, 0].plot(x, y3)
axs[1, 0].set_ylim(-10, 10)  #     
axs[1, 0].set_title('Tan(x)')

#  subplot (Exp(-x))
axs[1, 1].plot(x, y4)
axs[1, 1].set_title('Exp(-x)')

#    
plt.tight_layout()
plt.show()
```
